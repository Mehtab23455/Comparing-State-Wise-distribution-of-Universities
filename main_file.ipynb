```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/STATE_AND_TYPEWISE_UNIVERSITIES.csv")
```

```python
df['State Public University'] = df["State Public University"].fillna(df["State Public University"].mean())
```

```python
df
```

```python
df['State Private University'] = df["State Private University"].fillna(df["State Private University"].mean(), inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python


# Sort DataFrame by "State Public University" in descending order
df_sorted = df.sort_values(by="State Public University", ascending=False)

# Plot the Bar Chart
plt.figure(figsize=(12, 6))  # Increase figure size
plt.bar(df_sorted["State"], df_sorted["State Public University"], color="red")

# Formatting
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.xlabel("State")
plt.ylabel("Number of Public Universities")
plt.title("Public Universities by State (Sorted)")

# Show Plot
plt.show()
```

```python
# Sort DataFrame by "State Public University" in descending order
df_sorted = df.sort_values(by="State Private University", ascending=False)

# Plot the Bar Chart
plt.figure(figsize=(12, 6))  # Increase figure size
plt.bar(df_sorted["State"], df_sorted["State Private University"], color="blue")

# Formatting
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.xlabel("State")
plt.ylabel("Number of Private Universities")
plt.title("Private Universities by State (Sorted)")

# Show Plot
plt.show()
```

```python

x = np.arange(len(df["State"]))  # X-axis positions
width = 0.4  # Width of bars

plt.figure(figsize=(20, 6))
plt.bar(x - width/2, df["State Public University"], width, label="Public Universities", color="red")
plt.bar(x + width/2, df["State Private University"], width, label="Private Universities", color="blue")

# Formatting
plt.xticks(x, df["State"], rotation=45)  # Set x-axis labels
plt.xlabel("State")
plt.ylabel("Number of Universities")
plt.title("Public vs Private Universities by State (Grouped)")
plt.legend()

# Show Plot
plt.show()
```
